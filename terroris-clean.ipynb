```python
import numpy as np
import pandas as pd
import unicodecsv
import matplotlib.pyplot as plt

dir = "./"


records = pd.read_csv(dir+"globalterrorismdb_0617dist.csv",encoding='ISO-8859-1',usecols=[ 1,8, 13, 14, 100,103])

print len(records)
print records.head()



```

```python
lat = records.groupby(['iyear','country_txt']).agg('mean')['latitude'].reset_index()
lon = records.groupby(['iyear','country_txt']).agg('mean')['longitude'].reset_index()

kill = records.groupby(['iyear','country_txt']).agg('sum')['nkillter'].reset_index()
wound = records.groupby(['iyear','country_txt']).agg('sum')['nwoundte'].reset_index()
counts = records.groupby(['iyear','country_txt']).size().rename('counts').reset_index()



print lat.__class__
tt =lat.merge(lon, on=['iyear','country_txt'])
tt = tt.merge(kill, on=['iyear','country_txt'])
tt = tt.merge(wound, on=['iyear','country_txt'])
tt = tt.merge(counts, on=['iyear','country_txt'])
tt['severe'] = tt.fillna(0)['counts']+ tt.fillna(0)['nkillter']+tt.fillna(0)['nwoundte']
tt = tt.drop('counts', 1)
print len(tt)
tt = tt.sort(['iyear', 'severe'], ascending=[1, 1])
print tt
tt.to_csv("terror.csv")

```
